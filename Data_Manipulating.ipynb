#### Relationships between Weight, Horsepower, Displacement, and Miles per Gallon.

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Scatter3d(
    x=Car_frame['weight'],
    y=Car_frame['horsepower'],
    z=Car_frame['displacement'],
    mode='markers',
    marker=dict(
        size=4,
        color= Car_frame['mpg'],
        colorscale='Viridis',
        opacity=0.8, colorbar=dict(title='mpg')
    )
)])
fig.update_layout(scene=dict(
    xaxis_title='Weight',
    yaxis_title='Horsepower',
    zaxis_title='Displacement'
), title='Interactive 3D Scatter Plot')
fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=Car_frame, x='origin', y='mpg', showmeans=True, palette='Reds')
plt.xlabel('Country')
plt.ylabel('Miles per Gallon')
plt.title('Miles per Gallon Comparison between Country Origin')

# Add statistical annotations
ax = sns.pointplot(data=Car_frame, x='origin', y='mpg', join=False, dodge=0.8, color='black')
for xtick in ax.get_xticks():
    ax.text(xtick, Car_frame['mpg'].mean(), 'p > 0.05', ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=Car_frame, x='mpg', hue='origin', multiple='stack', kde=True, palette='Reds')
plt.xlabel('Miles per Gallon')
plt.title('Miles per Gallon Distributive Comparison between Country Origin')

plt.show()
```
